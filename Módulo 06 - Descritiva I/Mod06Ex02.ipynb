####  3) Faça um histograma da idade da mãe

**O histograma** é uma ferramenta gráfica que nos ajuda a entender a distribuição de frequência de uma variável contínua. Ele divide a faixa de valores dessa variável em intervalos e mostra quantas vezes os valores ocorrem em cada intervalo. As barras no gráfico representam esses intervalos e a altura das barras indica a frequência ou contagem de valores dentro de cada intervalo.


```python
# Definimos o tamnho da figura
# Criamos o histograma usando a coluna 'IDADEMAE' com os parâmetros de 15 para intervalos entre as barras e preto para bordas 
# Definimos o título como 'Histograma da Idade da Mãe'
# Definimos o eixo x como 'Idade da Mãe"
# Definimos o eixo y como 'Frequência'
# Exibimos o gráfico

plt.figure(figsize=(8, 6))
plt.hist(df['IDADEMAE'], bins=15, edgecolor='black')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```
